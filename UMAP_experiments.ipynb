```python
reducer = umap.UMAP()
```

```python
circle_data = circle[
    [
        "col1",
        "col2",
    ]
].values
scaled_circle_data = StandardScaler().fit_transform(circle_data)
```

```python
embedding = reducer.fit_transform(circle)
```

```python
plt.scatter(
    embedding[:, 0],
    embedding[:, 1],
)
plt.gca().set_aspect('equal', 'datalim')
plt.title("UMAP projection of the circle dataset", fontsize=12)
```
